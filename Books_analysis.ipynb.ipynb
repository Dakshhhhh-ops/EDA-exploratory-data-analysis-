### creating a new dataframe that has no negative values

```python
df=df[df["Publishing Year"]>1900]#(taking above 1900 for better analyze)
```

### removing null values 

```python
df.dropna(subset=['Book Name'],inplace=True)
```

### Histogram of publishing years

```python
plt.hist(df['Publishing Year'])
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.title('Distribution of Publishing Year')
```

### pie chart for book category(genre)

```python
df['genre'].value_counts()
```

```python
df['genre'].value_counts().plot(kind='pie',autopct='%2f')
```

### group by

```python
df.groupby('Author')['Book_average_rating'].sum().head().reset_index()
```

```python
df.groupby('Author')['Book_average_rating'].mean().sort_values(ascending=False)
```

### boxplot

```python
sns.boxplot(x=df['genre'],y=df['Book_ratings_count'])
```

### scatterplot

```python
sns.scatterplot(x=df['sale price'],y=df['units sold'])
```

### Pie chart of language_counts

```python
lanuage_counts=df['language_code'].value_counts().plot(kind='pie',autopct='%2f')
```

### groupbying publisher revenue

```python
df.groupby("Publisher ")['publisher revenue'].sum().sort_values(ascending=False)
```

### group by author rating and book rating count

```python
df.groupby("Author_Rating")["Book_ratings_count"].sum().sort_values(ascending=False)
```

### group by on language code

```python
df.groupby("language_code").size().sort_values(ascending=False)
```

### scatter plot

```python
sns.scatterplot(x=df['Book_average_rating'],y=df['Book_ratings_count'])
```

### boxplot of units sold for each author rating

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line",marker="o")
```
